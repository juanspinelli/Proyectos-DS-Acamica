## RESPUESTA

Creo que los datos que me ayudaran a trabajar el problema son:

**1. Superficie de la propiedad**: Ya que es un factor clave a la hora de asignarle un precio, a mayor superficie, mayor deberá ser el precio.

**2. Superficie cubierta de la propiedad**: Si la propiedad cuenta con espacios descubiertos, esta tendrá un mayor costo.

**3. Piso de la propiedad**: Usualmente, las propiedades que se encuentran a mayor altura tienen un mayor precio de venta.

**4. Ubicación de la propiedad**: El valor de una propiedad está ligado a su ubicación ya que una zona con una mayor demanda contribuye al aumento del valor de la propiedad. 

**5. Orientación de la propiedad**: Esta caracteristica va a determinar el valor de la propiedad ya que una orietación favorable mejora la ventilación y la luminosidad de la propiedad, y estos son factores que aumentan su valor. A su vez, una propiedad que, por ejemplo, tiene vista al mar, o a un parque/plaza, es propensa a tener un mayor costo.

**6. Tipo de Propiedad**: Dependiendo de si la propiedad es, por ejemplo, un PH, un lote vacío, un departamento o una casa su precio va a variar.

**7. Antiguedad y estado de la propiedad**: Una propiedad nueva, sin uso, tiene un precio mayor que una con uso. A su vez, esta caracteristica no es tan determinante, puesto que una propiedad antigua pero renovada también podría tener un alto costo. 

**8. Accesibilidad**: Una propiedad que cuenta con numerosos transportes públicos a su alrededor tendrá mayor costo.

**9. Comodidades de la propiedad**: Una propiedad que cuente con cocheras, pileta, gimnasio, sum o que tenga rampas o ascensores de acceso para personas con discapacidad tendrá mayor costo.

#### Comenzaremos haciendo un **modelo de regresión lineal a modo de benchmark**, para luego poder compararlo con los modelos que nos interesa entrenar.

#### La métrica de evaluación que utilizaré es el **RMSE**, ya que, al igual que con MSE, al evaluar a mi regresor con esta métrica se lo penalizará en mayor medida si la predicción se encuentra muy alejada del valor real, es decir, es sensible a valores atípicos. Pero, a diferencia del MSE, RMSE no altera la unidad original de mis datos. Además, el RMSE representa la **desviación estándar** de las **diferencias entre el valor predicho y el valor real**. Por lo cual, es una medida de cuan dispersos se encuentran los resíduos.

#### Comenzaré por hacer un train-test-split con 30% de variables para testeo.

```python
X = data[["rooms", "bedrooms", "bathrooms", "surface_total", "surface_covered"]]
y = data[["price"]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
modelos = ['Regresión lineal']

for i, model in enumerate([linear_model]):
    y_train_pred = model.predict(X_train).reshape(-1,1)
    y_test_pred = model.predict(X_test).reshape(-1,1)
    
    
    print(f'Modelo: {modelos[i]}')

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    print(f'Raíz del error cuadrático medio en Train: {rmse_train}')
    print(f'Raíz del error cuadrático medio en Test: {rmse_test}')
    
    plt.figure(figsize = (20,5))

    plt.subplot(1,2,1)
    sns.distplot(y_train - y_train_pred, bins = 20, label = 'train')
    sns.distplot(y_test - y_test_pred, bins = 20, label = 'test')
    plt.xlabel('errores')
    plt.legend()

    
    ax = plt.subplot(1,2,2)
    ax.scatter(y_test,y_test_pred, s =2)
    
    lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes]
    ]
    
    ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
    plt.xlabel('y (test)')
    plt.ylabel('y_pred (test)')
    
    plt.tight_layout()
    plt.show()
```
